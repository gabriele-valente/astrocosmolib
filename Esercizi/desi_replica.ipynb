```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import LambdaCDM
from astropy.cosmology import w0waCDM
```

```python
z=np.linspace(0,3,1000)
cmb=LambdaCDM(67.14,0.3169,1-0.3169,0)
desicmb=LambdaCDM(68.173,0.3027,1-0.3027,0)
w0wacmbdesidesi5=w0waCDM(H0=66.74, Om0=0.3191, Ode0=1-0.3191, w0=-0.752, wa=-0.86)
```

```python
rd=147.05
#Valori sperimentali
#redshift
z1=0.295 
z2=0.510 
z3=0.706 
z4=0.934 
z5=1.321 
z6=1.484 
z7=2.330
#distanza D_V/rd
y1=7.942*rd
y2=12.720*rd
y3=16.050*rd
y4=19.721*rd
y5=24.252*rd
y6=26.055*rd
y7=31.267*rd
#distanza D_M/rd
p2=13.588*rd
p3=17.351*rd
p4=21.576*rd
p5=27.601*rd
p6=30.512*rd
p7=38.988*rd
#distanza D_H/rd
h2=21.863*rd
h3=19.455*rd
h4=17.641*rd
h5=14.176*rd
h6=12.817*rd
h7=8.632*rd
```

```python

c = 299792.458

D_M=cmb.comoving_distance(z)
D_M2=desicmb.comoving_distance(z)
D_M3=w0wacmbdesidesi5.comoving_distance(z)

dm1=cmb.comoving_distance(z1)
dm2=cmb.comoving_distance(z2)
dm3=cmb.comoving_distance(z3)
dm4=cmb.comoving_distance(z4)
dm5=cmb.comoving_distance(z5)
dm6=cmb.comoving_distance(z6)
dm7=cmb.comoving_distance(z7)

D_H=c/cmb.H(z)
D_H2=c/desicmb.H(z)
D_H3=c/w0wacmbdesidesi5.H(z)

dh1=c/cmb.H(z1)
dh2=c/cmb.H(z2)
dh3=c/cmb.H(z3)
dh4=c/cmb.H(z4)
dh5=c/cmb.H(z5)
dh6=c/cmb.H(z6)
dh7=c/cmb.H(z7)

D_V=(z*D_M**2*D_H)**(1/3)
D_V2=(z*D_M2**2*D_H2)**(1/3)
D_V3=(z*D_M3**2*D_H3)**(1/3)

dv1=(z1*dm1**2*dh1)**(1/3)
dv2=(z2*dm2**2*dh2)**(1/3)
dv3=(z3*dm3**2*dh3)**(1/3)
dv4=(z4*dm4**2*dh4)**(1/3)
dv5=(z5*dm5**2*dh5)**(1/3)
dv6=(z6*dm6**2*dh6)**(1/3)
dv7=(z7*dm7**2*dh7)**(1/3) 
```

```python
plt.plot(z1,y1/dv1,'o',color='grey')
plt.plot(z2,y2/dv2,'o',color='grey')
plt.plot(z3,y3/dv3,'o',color='grey')
plt.plot(z4,y4/dv4,'o',color='grey')
plt.plot(z5,y5/dv5,'o',color='grey')
plt.plot(z6,y6/dv6,'o',color='grey')
plt.plot(z7,y7/dv7,'o',color='grey')

plt.semilogx(z,D_V/D_V,label='LambdaCDM: CMB',color='black')
plt.semilogx(z,D_V2/D_V,label='LambdaCDM: DESI CMB',color='purple')
plt.semilogx(z,D_V3/D_V,label='w0wa: CMB DESI DESI5(2)',color='red',linestyle='--')
plt.xlabel('Redshift')
plt.ylabel('D_V(z)/D_V(0)')
plt.title('Isotropic BAO')
plt.legend()
plt.show()
```

```python
#plt.plot(z1,p1/dm1,'o',color='grey')
plt.plot(z2,p2/dm2,'o',color='grey')
plt.plot(z3,p3/dm3,'o',color='grey')
plt.plot(z4,p4/dm4,'o',color='grey')
plt.plot(z5,p5/dm5,'o',color='grey')
plt.plot(z6,p6/dm6,'o',color='grey')
plt.plot(z7,p7/dm7,'o',color='grey')

plt.semilogx(z,D_M/D_M,label='LambdaCDM: CMB',color='black')
plt.semilogx(z,D_M2/D_M,label='LambdaCDM: DESI CMB',color='purple')
plt.semilogx(z,D_M3/D_M,label='w0wa: CMB DESI DESI5',color='red',linestyle='--')
plt.xlabel('Redshift')
plt.ylabel('D_M(z)/D_M(0)')
plt.title('Perpendicular BAO')
plt.legend()
plt.show()
```

```python
#plt.plot(z1,h1/dh1,'o',color='grey')
plt.plot(z2,h2/dh2,'o',color='grey')
plt.plot(z3,h3/dh3,'o',color='grey')
plt.plot(z4,h4/dh4,'o',color='grey')
plt.plot(z5,h5/dh5,'o',color='grey')
plt.plot(z6,h6/dh6,'o',color='grey')
plt.plot(z7,h7/dh7,'o',color='grey')

plt.semilogx(z,D_H/D_H,label='LambdaCDM: CMB',color='black')
plt.semilogx(z,D_H2/D_H,label='LambdaCDM: DESI CMB',color='purple')
plt.semilogx(z,D_H3/D_H,label='w0wa: CMB DESI DESI5',color='red',linestyle='--')
plt.xlabel('Redshift')
plt.ylabel('D_H(z)/D_H(0)')
plt.title('Parallel BAO')
plt.legend()
plt.show()
```
